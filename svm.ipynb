```python
from sklearn.datasets import fetch_20newsgroups

categories = ['alt.atheism', 'soc.religion.christian','comp.graphics', 'sci.med']
news = fetch_20newsgroups(subset='all', categories=categories)

data = news.data[:1000]
target = news.target[:1000]
target_names = news.target_names

print( "=== TARGET NAMES ===")
print( target_names )

print( "=== Data ===")
print( data[0] )
print( "=== Target ===")
print( target[0] )
print( "=== Target Name ===")
print( target_names[target[0]] )
```

# CountVectorizer Example

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    'This is the first document.',                #doc0
    'This is the second second document.',        #doc1
    'And the third one.',                         #doc2
    'Is this the first document?',                #doc3
    'The last document?',                         #doc4
]

# stop_words 인자를 통해 불용어 제거
vect = CountVectorizer(stop_words=['is', 'the'])
corpus_feature = vect.fit_transform(corpus).toarray()

doc1 =  corpus_feature[1]
print("{:^8} {:^8} {:^8} {:^8} {:^8} {:^8} {:^8} {:^8}".format(
    'and', 'document', 'first', 'last', 'one', 'second', 'the', 'third', 'this'))
print("{:^8} {:^8} {:^8} {:^8} {:^8} {:^8} {:^8} {:^8}".format(
        doc1[0], doc1[1], doc1[2], doc1[3], doc1[4], doc1[5], doc1[6],  doc1[7]))
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X = count_vect.fit_transform(data).toarray()

print(np.shape(X))

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2)

print(len(X_train), len(X_test))
```

```python
# Learning
from sklearn.svm import SVC
clf = SVC(kernel = 'linear')
clf.fit(X_train, y_train)

# Evaluation
y_pred = clf.predict(X_test)

from sklearn import metrics
print(metrics.classification_report(y_test, y_pred, target_names=target_names))
```

# 프로그래밍1

```python
import numpy as np
import operator
from collections import Counter
from sklearn.preprocessing import LabelEncoder
import copy


def countChange(data):
    return Counter(data.split(" "))
    
X,y=[],[]
f=open("C:/SMSSpamCollection")
data=list(line[:-1].split('\t') for line in f.readlines())

X=(list(data[i][1] for i in range(len(data))))
y=(list(data[i][0] for i in range(len(data))))

X = np.array(X)
y = np.array(y)

target_names=['ham','spam']

enc = LabelEncoder()
label_encoder = enc.fit(y)
y = label_encoder.transform(y)

data2=[]

for i in range(len(data)):        #각각의 데이터의 단어 개수
    data[i]=(countChange(X[i].lower()))

data2=Counter(data2)

for i in range(len(data)):   #단어 개수 총합
    data2=(data2+data[i])

sorted_x = sorted(dict(data2).items(),
               key=operator.itemgetter(1),reverse = True)
features=sorted_x[50:]     #상위 50개 제거
print (features)   #그 외의 단어들 출력

```

```python
#각각의 데이터에 상위 50개에 포함되는 단어 제거
ff=Counter()
for j in range(len(data)):
    ff=copy.deepcopy(data[j])
    for i in range(len(features)):
        if(features[i][0] in (data[j])):
            del (data[j][features[i][0]])
    data[j]=ff-data[j]
```

```python
#단어들을 복사해서 초기화하고
#각각의 데이터에 같은 단어가 있는 경우 값을 증가시킨다.
features2=[]
for i in range(len(features)):
    features2.append(features[i][0])
features2=Counter(features2)

date=[]
for i in range(len(data)):
    date.append(list(dict(features2+data[i]).values()))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(date, y, test_size=0.3)
print(len(X_train), len(X_test))
```

```python
# Learning
from sklearn.svm import SVC
clf = SVC(kernel = 'linear')
clf.fit(X_train, y_train)

# Evaluation
y_pred = clf.predict(X_test)

from sklearn import metrics
print(metrics.classification_report(y_test, y_pred, target_names=target_names))
```
